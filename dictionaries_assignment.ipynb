```python
d = {i: i*i for i in range(1, 11)}
even_dict = {key: value for key, value in d.items() if key % 2 == 0}
print(even_dict)
```

```python
from collections import defaultdict
d = defaultdict(list)
d['fruits'].append('apple')
d['fruits'].append('banana')
d['numbers'].append(10)
d['numbers'].append(20)
d['colors'].append('red')
print(d)
```

```python
def char_count(s):
    d = {}
    for ch in s:
        if ch in d:
            d[ch] += 1
        else:
            d[ch] = 1
    return d
text = "hello world"
result = char_count(text)
print(result)
```
