```python
matrix = []
print("Enter 9 values for a 3x3 matrix:")
for i in range(3):
    row = []
    for j in range(3):
        val = int(input(f"Enter element [{i+1},{j+1}]: "))
        row.append(val)
    matrix.append(tuple(row))   
matrix = tuple(matrix) 
print("\nMatrix:")
for row in matrix:
    print(row)
element = matrix[1][2]
print("\nElement at 2nd row, 3rd column:", element)
```

```python
def get_min(t):
    return min(t)
def get_max(t):
    return max(t)
def get_sum(t):
    return sum(t)
t = (5, 2, 8, 1, 9)
print("Tuple:", t)
print("Minimum:", get_min(t))
print("Maximum:", get_max(t))
print("Sum:", get_sum(t))
```
